```python
reset()
t=var('t')
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==x+4*y
deq2=diff(y,t)==x+y
syst=[deq1,deq2]
vars=[x,y]
in_cond=[0,1,2]
sol_syst=desolve_system(syst,vars,in_cond)
show(sol_syst)
sol_x(t)=sol_syst[0].rhs()
show(sol_x)
sol_y(t)=sol_syst[1].rhs()
show(sol_y)
g1=plot(sol_x(t),t,-4,4)
g2=plot(sol_y(t),t,-4,4,color='red')
g=g1+g2
g.show(xmin=-5,xmax=5,ymin=-5,ymax=5)
```
