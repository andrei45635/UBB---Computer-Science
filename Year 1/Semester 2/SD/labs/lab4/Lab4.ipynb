```python
t=var('t')
k=var('k')
x_0=var('x_0')
x=function('x')(t)
eqd=diff(x,t)==-k*x
#x(0)=x_0
sol(t,k,x_0)=desolve(eqd,[x,t],ivar=k,ics=[0,x_0], contrib_ode=True)
show(sol(t,k,x_0))
t12=var('t12')
sol_x=solve(sol(t12,k,x_0)==x_0/2,k)
sol_k=sol_x[0].substitute(t12==5730)
show(sol_k)
show(numerical_approx(sol_k.rhs(),digits=15))
sol_t=solve(sol(t,sol_k.rhs(),1)==1/5,t)
show(sol_t[0])
show(numerical_approx(sol_t[0].rhs()))
sol_e=solve(sol(t,sol_k.rhs(),0.91)==0.93,t)
show(numerical_approx(sol_e[0].rhs()))
g=plot(sol(t,1,1),-1,1)
for k in [2..10]:
    for x_0 in [1]:
        g1=plot(sol(t,k,x_0),-4,4,ymax=10,ymin=-1)
        g=g+g1
show(g)
```

```python
# Exercise 2
reset()
t=var('t')
k=var('k')
t_m=var('t_m')
t_0=var('t_0')
T=function('T')(t)
eqd=diff(T,t)==-k*(T-t_m)
sol(t,k,t_m,t_0)=desolve(eqd,[T,t],ivar=[k,t_m],ics=[0,t_0])
show('solutia problemei Cauchy:', sol(t,k,t_m,t_0))
sol1=solve(sol(t,k,21,36)==34.22,k)
show(sol1[0].rhs())
sol2=solve(sol(t+1,k,21,36)==34.11,k)
sol_new=solve(sol(1,k,21,34.22)==34.11,k)
sol_k=sol_new[0].rhs()
show(sol_k)
final_sol=solve(sol(t,sol_k,21,36)==34.22,t)
show('time:', find_root(final_sol[0], 0, 24))
show('time of death: 20:30 the previous day')
g=plot(sol(t,0,21,36),0,3)
for k in [1..10]:
    g1=plot(sol(t,k,21,36),0,3)
    g=g+g1
show(g)
```
